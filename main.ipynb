3. What percentage of total population survived?
 * Calculate and print the percentage of people who survived based on the data provided in the data set

```python
print('Total number of passangers who survivded : ', len(titanic[titanic['Survived'] == 1]))
print('Total number of passangers who died : ', len(titanic[titanic['Survived'] == 0]))
```

```python
f,ax=plt.subplots(1,2,figsize=(15,5))
titanic['Survived'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Survived')
ax[0].set_ylabel('')
sns.countplot('Survived',data=titanic,ax=ax[1])
ax[1].set_title('Survived')
ax[1].set_ylabel('Number of People')
plt.show()
```

```python
# Horizontal Bar plot for passenger survived or not
plt.figure(figsize=(5,5))
plt.barh(list(titanic['Survived'].value_counts().keys()),list(titanic['Survived'].value_counts()),color=['r','g'])
plt.title("Survival")
plt.xlabel('Number of People')
plt.ylabel('')
plt.show()
```

* According to abobe Tables and Graphs, we predict that the Survival rate [68%] of people is more than Death rate [32%] of people

```python
# Embarked

# only in titanic_df, fill the two missing values with the most occurred value, which is "S".
titanic["Embarked"] = titanic["Embarked"].fillna("S")

# plot
sns.factorplot('Embarked','Survived', data=titanic,size=4,aspect=3)

fig, (axis1,axis2,axis3) = plt.subplots(1,3,figsize=(15,5))

# sns.factorplot('Embarked',data=titanic,kind='count',order=['S','C','Q'],ax=axis1)
# sns.factorplot('Survived',hue="Embarked",data=titanic,kind='count',order=[1,0],ax=axis2)
sns.countplot(x='Embarked', data=titanic, ax=axis1)
sns.countplot(x='Survived', hue="Embarked", data=titanic, order=[1,0], ax=axis2)

# group by embarked, and get the mean for survived passengers for each value in Embarked
embark_perc = titanic[["Embarked", "Survived"]].groupby(['Embarked'],as_index=False).mean()
sns.barplot(x='Embarked', y='Survived', data=embark_perc,order=['S','C','Q'],ax=axis3)

# Either to consider Embarked column in predictions,
# and remove "S" dummy variable, 
# and leave "C" & "Q", since they seem to have a good rate for Survival.

# OR, don't create dummy variables for Embarked column, just drop it, 
# because logically, Embarked doesn't seem to be useful in prediction.

embark_dummies_titanic  = pd.get_dummies(titanic['Embarked'])
embark_dummies_titanic.drop(['S'], axis=1, inplace=True)

titanic = titanic.join(embark_dummies_titanic)

titanic.drop(['Embarked'], axis=1,inplace=True)
```

4. Calculate conditional probability
 * Calculate the probability that a passenger survived given that she is a female
Calculate the probability that a passenger survived given that he is a male

```python
# getting the frequency of number of passengers survived
titanic.groupby(['Sex','Survived'])['Survived'].count()
```

```python
#->  Display the probability that a passenger survived given that she/he is a female/male
titanic[["Sex","Survived"]].groupby(["Sex"], as_index = False).mean().sort_values(by="Survived",ascending = False)
```

```python
#->  Display the probability that a passenger survived given that she is a female
female = titanic.loc[titanic.Sex == 'female']["Survived"]
rate_female = sum(female)/len(female)

print("% of female who survived:", rate_female)
```

```python
male = titanic.loc[titanic.Sex == 'male']["Survived"]
rate_male = sum(male)/len(male)

print("% of male who survived:", rate_male)
```

```python
f,ax=plt.subplots(1,2,figsize=(16,7))
titanic['Survived'][titanic['Sex']=='male'].value_counts().plot.pie(explode=[0,0.2],autopct='%1.1f%%',ax=ax[0],shadow=True)
titanic['Survived'][titanic['Sex']=='female'].value_counts().plot.pie(explode=[0,0.2],autopct='%1.1f%%',ax=ax[1],shadow=True)
ax[0].set_title('Survived (male)')
ax[1].set_title('Survived (female)')

plt.show()
```

```python
sns.lineplot(x='Sex', y='Survived', data = titanic)
plt.title('Sex and Survivors')
plt.show()
```

* Accordind to above Tables and Graphs, we predict that the Survival rate of females [94%] is higher than Survival rate of Males [57%] and the Death rate of Females [6%] is also Less than the Death reatye of Male [43%]
